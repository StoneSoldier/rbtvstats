## Hauptbohnen
___

```python
import datetime

hosts = ('Budi', 'Eddy', 'Nils', 'Simon')
videoIds = videosHosts[videosHosts['host'].isin(hosts)].index.unique().tolist()
videosFiltered = videos[videos.index.isin(videoIds)]

#videosFiltered = videosFiltered[(videosFiltered['published'] >= datetime.datetime(2016, 10, 5))]
#videosFiltered = videosFiltered[(videosFiltered['published'] >= datetime.datetime(2016, 9, 21)) &
#                                (videosFiltered['published'] < datetime.datetime(2016, 9, 28))]

hostsLength = {
    'Budi': {},
    'Eddy': {},
    'Nils': {},
    'Simon': {}
}
for video in videosFiltered.iterrows():
    hostsContains = pd.Series(videosHosts.loc[video[0]]['host']).tolist()
    hostsContains = set(hostsContains).intersection(hosts)
    for host in hostsContains:
        month = video[1]['published'].strftime('%Y-%m')
        length = video[1]['length']
        if month not in hostsLength[host]:
            hostsLength[host][month] = 0
        hostsLength[host][month] = hostsLength[host][month] + (length.total_seconds() / 3600)

data = pd.DataFrame()
for host, monthly in hostsLength.iteritems():
    for month, length in monthly.iteritems():
        print (month, length)
        data.set_value(month, host, length)


data.sort_index(inplace=True)
data.plot(kind='box', figsize=(12, 6), title='Foobar', whis=100)
data.plot(kind='bar', figsize=(42, 16), title='Foobar')
data.mean()
```
